### Problem 2

### Approach

Even Fibonacci numbers are easily generated by the formula: `F_n+3 = 4 * F_n + F_n-3`

```python
def pe002(max:int=4000000) -> int:
    p0,p1=0,2
    total=0

    while (p1<max):
        total += p1
        p2 = 4*p1 + p0
        p0,p1 = p1,p2
    return total
```

### Problem 5

https://projecteuler.net/problem=5

### Approach

Factorize all number between 2 and 20, store the maximum exponents per prime, and multiply them to compose the least common multiple.

```python
def pe005(a:int=20) -> int:
    pri=[2,3,5,7,11,13,17,19]
    arr=[0]*len(pri)
    for i in range(2,a+1):
        k=i
        for j in range(len(pri)):
            c=0
            p=pri[j]
            while k%p==0:
                k //= p
                c += 1
            if arr[j] < c:
                arr[j] = c
    
    result = 1
    for i in range(len(arr)):
        result *= pri[i]**arr[i]
    
    return result
```
