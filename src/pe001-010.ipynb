Even Fibonacci numbers are easily generated by the formula: `F_n+3 = 4 * F_n + F_n-3`

````python
def pe002(max:int=4000000) -> int:
    p0,p1=0,2
    total=0

    while (p1<max):
        total += p1
        p2 = 4*p1 + p0
        p0,p1 = p1,p2
    return total
````

### Problem 5

https://projecteuler.net/problem=5

### Approach

Factorize all number between 2 and 20, store the maximum exponents per prime, and multiply them to compose the least common multiple.

````python
def pe005(a:int=20) -> int:
    pri=[2,3,5,7,11,13,17,19]
    arr=[0]*len(pri)
    for i in range(2,a+1):
        k=i
        for j in range(len(pri)):
            c=0
            p=pri[j]
            while k%p==0:
                k //= p
                c += 1
            if arr[j] < c:
                arr[j] = c
    
    result = 1
    for i in range(len(arr)):
        result *= pri[i]**arr[i]
    
    return result
````

### Problem 7

The main problem is to estimate where about the 10.001st prime would be found, in order to generate enough but not too many primes. Based on the [prime-counting function](https://en.wikipedia.org/wiki/Prime-counting_function) I put the upperbound safely at around 11.000. The rest is just a simple prime sieve.

````python
def pe007(num:int=10001):
    primax=num*11
    
    pri=[0]*primax
    
    for i in range(2, primax):
        if pri[i]==0:
            for j in range(i*i, primax, i):
                pri[j]=1
    
    c=0
    for i in range(2,primax): 
        if pri[i]==0:
            c += 1
            if c==num:
                return i
    return -1

pe007()
````
